````python
from langchain_ollama import OllamaLLM
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
````

````python
model = OllamaLLM(model="qwen2.5-coder:7b")

# 2. Création du template de prompt
prompt = ChatPromptTemplate.from_template(
    """
You are a Senior Software Architect specializing in Python. Your goal is to provide the most readable, PEP 8 compliant, and maintainable solution possible.

## CRITICAL INSTRUCTION:
DO NOT use Markdown formatting. DO NOT use backticks (
```). DO NOT include any text before or after the code. Your entire response must be valid Python code that can be saved directly to a .py file and executed.

### OUTPUT STRUCTURE:
1. Valid Python code only.
2. At the very end of the file, provide a DESIGN REFLECTION inside a Python comment block.

## REQUEST
{topic}
""")

# 3. Construction de la chaîne (Chain)
# On ajoute StrOutputParser() pour être sûr de récupérer du texte brut
chain = prompt | model | StrOutputParser()

# 4. Exécution
topic_input = """Generate a REST API helloworld. 
Use the stack FastAPI.
Save the file as project2.py"""
print("Génération en cours...")
response = chain.invoke({"topic": topic_input})

# 5. Sauvegarde dans un fichier
file_name = "project.py"
with open(file_name, "w", encoding="utf-8") as f:
    f.write(response)

print(f"✅ La réponse a été sauvegardée dans {file_name}")
````
